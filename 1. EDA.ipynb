# IV. Correlations

 >Relationships between numeric features and other numeric features.
 
 >Correlation is a value between -1 and 1 that represents how closely values for two separate features move in unison.

```python
# Calculate correlations between numeric features
correlations = df.corr()
```

```python
correlations
```

It's important to notice that the correlations for 'basement' all show as NaN. This is expected because right now that feature doesn't vary at all (its standard deviation is 0).

###### Let's try to visualize the correlation grid and make it easier to digest.

Seaborn's helpful sns.heatmap() function does the trick.

>The cmap= argument controls the color palette used in the heatmap. We set it to the value of RdBu_r, which is a diverging red-blue color map. So positive correlation values will be red and negative correlation values will be blue.

```python
# Increase the figsize to 7 x 6
plt.figure(figsize=(7,6))

# Plot heatmap of correlations
sns.heatmap(correlations, cmap='RdBu_r')
```

Perhaps the most helpful way to interpret this correlation heatmap is to start by scanning the left side and find the row with our target variable, which is 'tx_price'.

>Then, scan across to see if any of the boxes are dark red (strong positive correlation) or dark blue (strong negative correlation).

>In this case, it doesn't look like many features are strongly correlated with the target variable.

>There is a weak positive correlation between 'tx_price' and 'property_tax'.

###### Next, when plotting a heatmap of correlations, it's often helpful to do four things:

>Change the background to white. This way, 0 correlation will show as white

>Annotate the cell with their correlations values

>Mask the top triangle (less visual clutter)

>Drop the legend (colorbar on the side)

```python
# Change color scheme
sns.set_style("white")

# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of annotated correlations
sns.heatmap(correlations * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f')

plt.show()
```

So one thing to notice about the correlation matrix is that the upper triangle is an exact mirror of the lower triangle. This is because all of the features appear on both the x-axis AND the y-axis. In other words, the grid will show both the correlation between "A and B" as well the correlation between "B and A".

However, the correlation between "A and B" is exactly the same as the correlation between "B and A"! 

###### Generate a boolean mask for the top triangle.

The mask should have the same shape (i.e. dimensions) as the correlation matrix.

The values in the upper triangle should be True/1 while those in the lower triangle should be False/0.

Either a mask of 0's and 1's OR a mask of False's and True's will work.

Tip: np.zeros_like() returns an array of zeros with the same shape as a given array.

Tip: np.triu_indices_from() returns the indices for the upper triangle of an array.

Tip: Remember that NumPy arrays are mutable, which means that you can "overwrite" the values in any array by selecting those values by their indices.
Tip: Feel free to insert more code cells and play around with this a bit. This is meant to be a trickier exercise, so check the answer key if the solution still eludes you after several minutes of trial-and-error.

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations)
mask[np.triu_indices_from(mask)] = 1

# Option 2: Mask of False/True
# mask = np.zeros_like(correlations, dtype=np.bool)
# mask[np.triu_indices_from(mask)] = True
```

```python
sns.heatmap( mask )
plt.show()
```

```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(correlations * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f',
            mask=mask)
plt.show()
```

```python
# Make the figsize 9 x 8
plt.figure(figsize=(9,8))

# Plot heatmap of correlations
sns.heatmap(correlations * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f',
            mask=mask,
            cbar=False)
plt.show()
```
